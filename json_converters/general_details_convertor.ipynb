```python
import pandas as pd
import json
```

```python
# Load Excel file
file_path = "../excel_files/General_details.xlsx"
df = pd.read_excel(file_path)
df = df.fillna("")
df
```

```python
# Split 'Languages' and 'Language_level' into lists
df.loc[df['Detail'] == 'Languages', 'Value'] = df.loc[df['Detail'] == 'Languages', 'Value'].str.split(', ')
df.loc[df['Detail'] == 'Language_level', 'Value'] = df.loc[df['Detail'] == 'Language_level', 'Value'].str.split(', ')

# Display the modified DataFrame
df
```

```python
json_data = pd.Series(df['Value'].values, index=df['Detail']).to_dict()
json_data
```

```python
import json
from datetime import datetime

class DateOnlyEncoder(json.JSONEncoder):
    def default(self, obj):
        if isinstance(obj, datetime):
            # Format as "Month Day, Year" (e.g., "November 13, 1990")
            # return obj.strftime("%B %d, %Y")
            return obj.strftime("%B %d")
        return super().default(obj)


with open("../json_files/general_details.json", "w", encoding="utf-8") as json_file:
    json.dump(json_data, json_file, indent=4, cls=DateOnlyEncoder)
```
