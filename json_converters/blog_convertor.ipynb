```python
import pandas as pd
```

```python
# Load Excel file
file_path = "../excel_files/Blog_details.xlsx"
df = pd.read_excel(file_path)
df = df.fillna("")
df
```

```python
# Convert DataFrame to JSON
json_data = df.to_dict(orient="records")
json_data
```

```python
import json
from datetime import datetime

class DateOnlyEncoder(json.JSONEncoder):
    def default(self, obj):
        if isinstance(obj, datetime):
            # Format as "Month Day, Year" (e.g., "November 13, 1990")
            return obj.strftime("%B %d, %Y")
        return super().default(obj)

with open("../json_files/blog_details.json", "w", encoding="utf-8") as json_file:
    json.dump(json_data, json_file, indent=4, cls=DateOnlyEncoder)
```
