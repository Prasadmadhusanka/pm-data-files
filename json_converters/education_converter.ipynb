```python
import pandas as pd
import json
```

```python
# Load Excel file
file_path = "../excel_files/Education_details.xlsx"
df = pd.read_excel(file_path)
df = df.fillna("")
df
```

```python
# Function to convert key_modules column into a list
def convert_modules(modules_str):
    if isinstance(modules_str, str):
        return [modules.strip() for modules in modules_str.split(",")]  # Split by comma and trim spaces
    return []
```

```python
# Apply conversion function to 'key_modules' column
df["key_modules"] = df["key_modules"].apply(convert_modules)
```

```python
# Convert DataFrame to JSON
json_data = df.to_dict(orient="records")
json_data
```

```python
with open("../json_files/education_details.json", "w", encoding="utf-8") as json_file:
    json.dump(json_data, json_file, indent=4)
```
