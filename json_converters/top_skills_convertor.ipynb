```python
import pandas as pd
import json
```

```python
# Load Excel file
file_path = "../excel_files/Skills_details.xlsx"
df = pd.read_excel(file_path, sheet_name='top_skills')
df
```

```python
df.columns.values
```

```python
# Convert DataFrame to JSON
json_data = df.to_dict(orient="records")
json_data
```

```python
# Save to JSON file
with open("../json_files/top_skills_details.json", "w", encoding="utf-8") as json_file:
    json.dump(json_data, json_file, indent=4)
```
