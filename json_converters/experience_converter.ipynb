```python
import pandas as pd
import json
```

```python
# Load Excel file
file_path = "../excel_files/Experience_details.xlsx"
df = pd.read_excel(file_path)
df = df.fillna("")
df
```

```python
# Function to convert key_modules column into a list
def convert_technology(technology_str):
    if isinstance(technology_str, str):
        return [technology.strip() for technology in technology_str.split(",")]  # Split by comma and trim spaces
    return []
```

```python
# Apply conversion function to 'key_modules' column
df["technology_used"] = df["technology_used"].apply(convert_technology)
```

```python
# Function to convert responsibility column into a list
def convert_responsibility(responsibility_str):
    if isinstance(responsibility_str, str):
        return [responsibility.strip() for responsibility in responsibility_str.split("|")]  # Split by comma and trim spaces
    return []
```

```python
# Apply conversion function to 'responsibility' column
df["responsibility"] = df["responsibility"].apply(convert_responsibility)
```

```python
# Convert DataFrame to JSON
json_data = df.to_dict(orient="records")
json_data
```

```python
with open("../json_files/experience_details.json", "w", encoding="utf-8") as json_file:
    json.dump(json_data, json_file, indent=4)
```
