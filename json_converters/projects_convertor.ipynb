```python
import pandas as pd
import json
```

```python
# Load Excel file
file_path = "../excel_files/Projects_details.xlsx"
df = pd.read_excel(file_path)
df = df.fillna("")
df.head()
```

```python
# Function to convert skills column into a list
def convert_skills(skill_str):
    if isinstance(skill_str, str):
        return [skill.strip() for skill in skill_str.split(",")]  # Split by comma and trim spaces
    return []
```

```python
# Apply conversion function to 'skills' column
df["technologies"] = df["technologies"].apply(convert_skills)
```

```python
# Function to convert category column into a list
def convert_categories(category_str):
    if isinstance(category_str, str):
        return [category.strip() for category in category_str.split(",")]  # Split by comma and trim spaces
    return []
```

```python
# Apply conversion function to 'category' column
df["category"] = df["category"].apply(convert_categories)
```

```python
# Convert DataFrame to JSON
json_data = df.to_dict(orient="records")
json_data
```

```python
import json
from datetime import datetime

class DateOnlyEncoder(json.JSONEncoder):
    def default(self, obj):
        if isinstance(obj, datetime):
            # Format as "Month Year" (e.g., "November 1990")
            return obj.strftime("%B %Y")
        return super().default(obj)


with open("../json_files/project_details.json", "w", encoding="utf-8") as json_file:
    json.dump(json_data, json_file, indent=4, cls=DateOnlyEncoder)
```
