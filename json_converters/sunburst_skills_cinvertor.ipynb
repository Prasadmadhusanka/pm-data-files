```python
import pandas as pd
import json
```

```python
# Load Excel file
file_path = "../excel_files/skills_details.xlsx"
df = pd.read_excel(file_path, sheet_name='Sheet3')
df = df.fillna("")
df
```

```python
df['ids'] = df['ids'].str.strip().str.replace(" ", "", regex=False)
df['parents'] = df['parents'].str.strip().str.replace(" ", "", regex=False)
```

```python
visible_df = df[df['visibility'] == True]
visible_df
```

```python
visible_df = visible_df[["ids", "labels", "parents"]]
visible_df
```

```python
# Convert DataFrame to JSON
json_data = visible_df.to_dict(orient="records")
json_data
```

```python
# Save to JSON file
with open("../json_files/sunburst_details.json", "w", encoding="utf-8") as json_file:
    json.dump(json_data, json_file, indent=4)
```
